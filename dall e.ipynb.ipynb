```python

# Install TensorFlow and import necessary libraries
!pip install tensorflow
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from skimage.transform import resize

# Load the CIFAR-10 dataset
(train_images, train_labels), (_, _) = cifar10.load_data()

# Class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Function to display images
def display_images(images, labels, class_names, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(15, 15))
    for i in range(num_images):
        axes[i].imshow(images[i])
        axes[i].set_title(class_names[labels[i][0]])
        axes[i].axis('off')
    plt.show()

# Get indices of cat and dog images
cat_indices = np.where(train_labels == 3)[0]  # 3 is the index of the 'cat' class
dog_indices = np.where(train_labels == 5)[0]  # 5 is the index of the 'dog' class

# Select random cat and dog images
random_cat_indices = np.random.choice(cat_indices, 10, replace=False)  # Selecting 10 random cat images
random_dog_indices = np.random.choice(dog_indices, 10, replace=False)  # Selecting 10 random dog images

# Resize images to 64x64 pixels
resized_cat_images = np.array([resize(train_images[idx], (64, 64)) for idx in random_cat_indices])
resized_dog_images = np.array([resize(train_images[idx], (64, 64)) for idx in random_dog_indices])

# Display resized cat and dog images
display_images(resized_cat_images, train_labels[random_cat_indices], class_names, num_images=10)
display_images(resized_dog_images, train_labels[random_dog_indices], class_names, num_images=10)
```
